```python
import tweepy

import pandas as pd
```

```python
# read auth details from text file

auth_details_read = open("C:\\Users\\Q\\Desktop\\Cypress\\twit_auth.txt","r")
```

```python
auth_details = auth_details_read.readlines()

auth_details_read.close()
```

```python
# auth_details
```

```python
CONSUMER_KEY = auth_details[0].strip() # INSERT API/CONSUMER KEY
CONSUMER_SECRET = auth_details[1].strip() # INSERT API/CONSUMER SECRET
ACCESS_TOKEN = auth_details[2].strip() # INSERT ACCESS TOKEN
ACCESS_TOKEN_SECRET = auth_details[3] # INSERT ACCESS SECRET
```

```python
# PASS IN YOUR AUTHENTICATION DETAILS

auth = tweepy.OAuth1UserHandler(CONSUMER_KEY,
                                CONSUMER_SECRET,
                                ACCESS_TOKEN,
                                ACCESS_TOKEN_SECRET)
```

```python
# INSTANTIATE

api = tweepy.API(auth)
```

```python
# GET TWEETS FROM MY TIMELINE

home_tweets = api.home_timeline(tweet_mode = "extended")
```

```python
tweet_id = []
created_at = []
user_name = []
name = []
tweet_text = []
retweeted_by = []
hashtags = []
```

```python
for tweet in home_tweets:
    if 'RT' in tweet.full_text:
        tweet_id.append(tweet.retweeted_status.id)

        created_at.append(str(tweet.retweeted_status.created_at.year) + '-' + 
                          str(tweet.retweeted_status.created_at.month) + '-' + 
                          str(tweet.retweeted_status.created_at.day))

        user_name.append(tweet.retweeted_status.user.screen_name)

        name.append(tweet.retweeted_status.user.name)

        tweet_text.append(tweet.retweeted_status.full_text)

        retweeted_by.append(tweet.user.screen_name)

        if len(tweet.entities['hashtags']) > 0:
            for hashtag in tweet.entities['hashtags']:
                hashtags.append(hashtag)
        else:
            hashtags.append('#')
    else:
        tweet_id.append(tweet.id)
        
        created_at.append(str(tweet.created_at.year) + '-' + 
                          str(tweet.created_at.month) + '-' + 
                          str(tweet.created_at.day))
        
        user_name.append(tweet.user.screen_name)
        
        name.append(tweet.user.name)
        
        tweet_text.append(tweet.full_text)
        
        retweeted_by.append('')
        
        if len(tweet.entities['hashtags']) > 0:
            for hashtag in tweet.entities['hashtags']:
                hashtags.append(hashtag)
        else:
            hashtags.append('#')
```

```python
tweets_scrape_to_dict = {'Tweet_ID': tweet_id,
                        'User_Name': user_name,
                        'Name': name,
                        'Tweet_Text': tweet_text,
                        'Retweeted_By': retweeted_by,
                        'Hashtags': hashtags
                       }
```

```python
tweets_scrape_df = pd.DataFrame(tweets_scrape_to_dict)
```
